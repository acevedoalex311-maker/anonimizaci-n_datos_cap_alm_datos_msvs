## Importación de librerías

```python
import kagglehub
import pandas as pd
import os
import glob
import hashlib
import numpy as np
```

## Llamado del df Social Media 

```python
# download the latest version – returns the path to the downloaded files
path = kagglehub.dataset_download("anoopjohny/socialmedia")
print("Path to dataset files:", path)

if os.path.isdir(path):
    csv_files = glob.glob(os.path.join(path, "*.csv"))
else:
    # if path ends in .zip you can unzip it yourself
    import zipfile
    with zipfile.ZipFile(path) as z:
        z.extractall(os.path.dirname(path))
    csv_files = glob.glob(os.path.join(os.path.dirname(path), "*.csv"))

print("Found CSV files:", csv_files)

# load one (or more) into a DataFrame
df = pd.read_csv(csv_files[0])       
df.head()
```

### Copia del df para proceso de anonimización

```python
social_df = df.copy()
```

```python
# --- 1. SEUDONIMIZACIÓN (Identificadores hash) ---
def pseudonymize(val):
    if pd.isna(val):
        return val
    # Create a 12-character SHA-256 hash for consistency and privacy
    return hashlib.sha256(str(val).encode()).hexdigest()[:12]

social_df['User ID'] = social_df['User ID'].apply(pseudonymize)
social_df['Username'] = social_df['Username'].apply(pseudonymize)
social_df['Post ID'] = social_df['Post ID'].apply(pseudonymize)


# --- 2. GENERALIZACIÓN (Reducción de la precisión) ---
def generalize_location(loc):
    # Reducción de la localidad a nivel de país o región
    if pd.isna(loc) or ',' not in loc:
        return loc
    return loc.split(',')[-1].strip()

def generalize_date(date_str, level='year'):
    # Agrupación de fechas a nivel de año o mes
    try:
        dt = pd.to_datetime(date_str, dayfirst=True)
        if level == 'year':
            return str(dt.year)
        elif level == 'month':
            return dt.strftime('%Y-%m')
    except:
        return date_str

def bin_count(val):
    # Agrupado en bins 
    if val < 1000: return '<1K'
    elif val < 10000: return '1K-10K'
    elif val < 50000: return '10K-50K'
    else: return '50K+'

social_df['Location'] = social_df['Location'].apply(generalize_location)
social_df['Server Post'] = social_df['Server Post'].apply(generalize_location)
social_df['Post Timestamp'] = social_df['Post Timestamp'].apply(lambda x: generalize_date(x, 'month'))
social_df['Account Creation Date'] = social_df['Account Creation Date'].apply(lambda x: generalize_date(x, 'year'))
social_df['User Followers Range'] = social_df['User Followers'].apply(bin_count)
social_df['User Following Range'] = social_df['User Following'].apply(bin_count)


# --- 3. INTERRUPCIÓN (Añadir ruido aleatorio) ---
def add_noise(val, percentage=0.05):
    # Adicionar ruido gaussiano a los valores numéricos para evitar la identificación directa
    if pd.isna(val) or not isinstance(val, (int, float)):
        return val
    noise = np.random.normal(0, val * percentage)
    return int(max(0, val + noise))

numerical_cols = ['Likes/Reactions', 'Comments', 'Shares/Retweets', 'User Engagement', 'User Interactions']
for col in numerical_cols:
    social_df[col] = social_df[col].apply(add_noise)

# --- 4. ENMASCARAMIENTO (Ocultación de URL) ---
social_df['Media URL'] = social_df['Media URL'].apply(lambda x: 'http://[REDACTED]' if pd.notna(x) else x)
social_df['Post URL'] = social_df['Post URL'].apply(lambda x: 'http://[REDACTED]' if pd.notna(x) else x)


# --- 5. ELIMINACIÓN (eliminación de datos no estructurados de alto riesgo) ---
# Eliminar campos que probablemente contengan PII específica o prosa identificable
cols_to_delete = ['Post Text', 'User Bio', 'User Description 1', 'User Description 2', 'Mentions']
df_anon = social_df.drop(columns=cols_to_delete)

# Guardado del dataset anonimizado
df_anon.to_csv('anonymized_socialmedia.csv', index=False)

print("Anonymization complete. File saved as 'anonymized_socialmedia.csv'.")
print(df_anon.head())
```
